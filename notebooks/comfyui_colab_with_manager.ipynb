### Run ComfyUI with Ngrok (Colab localhost)




```python
import subprocess
import threading
import time
import socket
import urllib.request
import json
import os

# 1. Asegurar ruta absoluta de trabajo
%cd /content/ComfyUI

# 2. Configurar Token de Ngrok de forma global
NGROK_TOKEN = "3GeYWdsVDY4LHfOEzKk3CjAi8Sr_49PRkiWwEJSs1z3uxXXW5"
os.system(f"ngrok config add-authtoken {NGROK_TOKEN}")

# Limpiar cualquier proceso previo colgado
os.system("killall ngrok 2>/dev/null")

PORT = 8188

def obtener_y_mostrar_enlace(port):
    # Esperar a que el puerto interno 8188 de ComfyUI responda
    while True:
        time.sleep(1)
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        result = sock.connect_ex(('127.0.0.1', port))
        sock.close()
        if result == 0:
            break

    time.sleep(2)

    # Lanzar el túnel HTTP de Ngrok en segundo plano
    subprocess.Popen(["ngrok", "http", str(port)], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

    # Esperar un momento a que Ngrok negocie la URL con sus servidores
    time.sleep(3)

    # Consultar la API local de Ngrok para extraer la URL pública real
    try:
        with urllib.request.urlopen('http://127.0.0.1:4040/api/tunnels') as response:
            data = json.loads(response.read().decode())
            url = data['tunnels'][0]['public_url']
            print("\n" + "="*70)
            print(f"🌍 ENLACE DEFINITIVO DE ACCESO A COMFYUI:")
            print(f"👉 {url}")
            print("="*70 + "\n")
    except Exception as e:
        print("\n❌ [ERROR] Ngrok no pudo entregar la URL. Revisa tu panel en ngrok.com")

# Lanzar el hilo encargado de capturar y mostrar la URL
threading.Thread(target=obtener_y_mostrar_enlace, daemon=True, args=(PORT,)).start()

print("🟢 Iniciando motor de ComfyUI... Tu enlace aparecerá abajo en unos segundos.")

# 3. Arrancar ComfyUI abriendo los encabezados CORS para evitar el error 403
!python /content/ComfyUI/main.py --dont-print-server --listen 0.0.0.0 --port {PORT} --enable-cors-header "*"
```

### Run ComfyUI with colab iframe (use only in case the previous way with localtunnel doesn't work)

You should see the ui appear in an iframe. If you get a 403 error, it's your firefox settings or an extension that's messing things up.

If you want to open it in another window use the link.

Note that some UI features like live image previews won't work because the colab iframe blocks websockets.

```python
import threading
import time
import socket
def iframe_thread(port):
  while True:
      time.sleep(0.5)
      sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
      result = sock.connect_ex(('127.0.0.1', port))
      if result == 0:
        break
      sock.close()
  from google.colab import output
  output.serve_kernel_port_as_iframe(port, height=1024)
  print("to open it in a window you can open this link here:")
  output.serve_kernel_port_as_window(port)

threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

!python main.py --dont-print-server
```
